```python
#(5)histogram of sepal length distribution
```

```python
plt.hist(data['SL'],color='brown')
plt.title('Distribution of Sepal Length(SL)',color='green')
plt.xlabel('SL')
plt.ylabel('Freuency')
plt.show()
```

```python
#scatter plot of Sepal length and sepal width
data.plot(kind='scatter', x='SL',y='SW')
plt.title('Sepal Length(SL) vs Sepal Width(SW)')
plt.show()
```

```python
#box plot for Sepal Length

plt.boxplot(data['SL'])
plt.show()
```

```python
#boxplot for Sepal and Petal data
plt.boxplot([data['SL'],data['SW'],data['PL'],data['PW']])
plt.show()
```

```python
#scatter plot of sepal data 
colors = {'Iris-setosa':'red','Iris-virginica':'blue','Iris-versicolor':'green'}
plt.figure(figsize=(12,8))
plt.scatter(data['SL'], data['SW'], c=data['Classification'].map(colors))
plt.title('Sepal Scatter Plot')
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

```python
#pairplot using seaborn
sns.pairplot(data)
plt.show()
```

```python
data.boxplot(column=['SL'], by=['Classification'])
plt.show()
```
